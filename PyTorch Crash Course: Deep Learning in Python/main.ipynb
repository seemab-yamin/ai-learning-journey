## Automatic Differentiation

```python
a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)

f = 3 * a**3 - b**2
f
```

```python
# f.sum(), f.sum().backward(), f, f.sum()
```

```python
f.backward(gradient=torch.tensor([1, 1]))
```

```python
a.grad
```

```python
9 * a**2
```

```python
b.grad
```
