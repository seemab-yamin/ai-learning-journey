## DataLoader in PyTorch

```python
from torch.utils.data import DataLoader, TensorDataset

# for loading dataset
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# load the dataset from sklearn
X, y = load_breast_cancer(return_X_y=True)
X[0], y[0]
```

```python
# split 20% test data for testing the model after training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# scaling ensures all features contribute equally
# normalized feature vectors: have better performance in neural networks
scaler = StandardScaler()

# learn mean/std on training data
X_train_scaled = scaler.fit_transform(X_train)

# apply the learned mean/std from training data here
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled[0], X_test_scaled[0]
```

```python
X_train_scaled_tensor = torch.from_numpy(X_train_scaled).float()
X_test_scaled_tensor = torch.from_numpy(X_test_scaled).float()

y_train_tensor = torch.from_numpy(y_train).float().unsqueeze(1) # make it a column vector
y_test_tensor = torch.from_numpy(y_test).float().unsqueeze(1) # make it a column vector
```

```python
X_train_scaled_tensor.shape, X_test_scaled_tensor.shape, y_train_tensor.shape, y_test_tensor.shape
```

```python
train_dataset = TensorDataset(X_train_scaled_tensor, y_train_tensor)
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

## Build Neural Network Model

```python
import torch.nn as nn
import torch.nn.functional as F # activation function
import torch.optim as optim # for optimizer

# define neural network architecture
class BCNet(nn.Module):
    def __init__(self):
        super(BCNet, self).__init__()

        self.fcl = nn.Linear(
            30, 64
        )  # input features = 30, output features = 64 will be given to the next layer
        self.fcl2 = nn.Linear(
            64, 32
        )  # input features = 64, output features = 32 will be given to the next layer
        self.fcl3 = nn.Linear(
            32, 1
        )  # input features = 32, output features = 1 will be the actual output of the model (probability of being malignant)

    def forward(self, x):
        x = F.relu(self.fcl(x))  # relu activation function to introduce non-linearity
        x = F.relu(self.fcl2(x))  # relu activation function to introduce non-linearity
        x = F.sigmoid(
            self.fcl3(x)
        )  # sigmoid to get output between 0 and 1 (probability)

        return x
```

```python
model = BCNet()
```

```python
criterion = (
    nn.BCELoss()
)  # binary cross entropy loss for binary classification - loss function to optimize
optimizer = optim.Adam(
    params=model.parameters(), lr=0.001
)  # Adam optimizer with learning rate of 0.001 - optimization algorithm to update weights
```

```python
# training loop
epochs = 20

for epoch in range(epochs):
    model.train()  # set model to training mode
    running_loss = 0.0
    
    for x_batch, y_batch in train_loader:
        optimizer.zero_grad()  # zero the gradients before backpropagation

        preds = model(x_batch)  # forward pass to get predictions
        loss = criterion(preds, y_batch)  # calculate loss
        
        loss.backward()  # backpropagation to calculate gradients
        
        optimizer.step()  # update weights based on gradients
        
        running_loss += loss.item() # accumulate loss for the epoch
    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader):.4f}")
```

## Evaluation on test set

```python
with torch.no_grad():  # no need to calculate gradients for evaluation
    model.eval()  # set model to evaluation mode

    preds = model(X_test_scaled_tensor)  # get predictions for test data
    
    loss = criterion(preds, y_test_tensor)  # calculate loss on test data
    print(f"Test Loss: {loss.item():.4f}")
    
    accuracy = (( preds > 0.5) == y_test_tensor).float().mean().item() # calculate accuracy by comparing predictions to true labels
    print(f"Test Accuracy: {accuracy:.4f}")
```
